```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
# criando o dataframe
df = pd.read_excel('./data_set/AdventureWorks.xlsx')
```

```python
# Visualizando as 5 primeiras linhas
df.head()
```

```python
# Verificando qtde de linhas e colunas
df.shape
```

```python
# verificando os tipos de dados
df.dtypes
```

```python
# Valor total de receitas
df['Valor Venda'].sum()
```

```python
# Custo Venda
df['Custo'] = df['Custo Unitário'] * df['Quantidade']
df.head()
```

```python
# Custo Total
df['Custo'].sum()
```

```python
# Descobrindo o Lucro
df['Lucro'] = df['Valor Venda'] - df['Custo']
df.head()
```

```python
# Total Lucro
round(df['Lucro'].sum(),2)
```

```python
# Criando coluna com o tempo para envio em dias
df['Tempo Envio'] = df['Data Envio'] - df['Data Venda']
df.head()
```

## Tempo Medio do Envio por Marcas

```python
# Extraindo apenas os dias do tempo de envio
df['Tempo Envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
df.head()
```

```python
# Media do tempo de envio por Marca
df.groupby('Marca')['Tempo Envio'].mean()
```

## Missing Values

```python
# Verificando se temos dados faltantes
df.isnull().sum()
```

## Descobrindo o lucro por Ano e Marca

```python
# Agrupando por Ano e Marca
pd.options.display.float_format = '{:20,.2f}'.format
df.groupby([df['Data Venda'].dt.year, df['Marca']])['Lucro'].sum()
```

```python
# Lucro Ano e Marca
lucro_ano = df.groupby([df['Data Venda'].dt.year, df['Marca']])['Lucro'].sum().reset_index()
lucro_ano.head()
```

```python
# Total de produtos vendidos
df.groupby([df['Produto']])['Quantidade'].sum().sort_values(ascending = False)
```

```python
# Grafico com total de produto vendidos
df.groupby([df['Produto']])['Quantidade'].sum().sort_values(ascending = True).plot.barh(title = 'Total Produtos Vendidos')
plt.xlabel('Total')
plt.ylabel('Produto')
```

```python
# Grafico com Lucro por Ano
df.groupby([df['Data Venda'].dt.year])['Lucro'].sum().plot.bar(title = 'Lucro x Ano')
plt.xlabel('Ano')
plt.ylabel('Receita')
```

```python
# Selecionando as venda de 2009
df_vendas2009 = df.loc[(df['Data Venda'].dt.year == 2009)]
df_vendas2009
```

```python
# Grafico com Lucro por mes em 2009
df.groupby([df['Data Venda'].dt.month])['Lucro'].sum().plot(title = 'Lucro x Mes')
plt.xlabel('Mês')
plt.ylabel('Receita')
```

```python
# Grafico com Lucro por Marca em 2009
df.groupby([df['Marca']])['Lucro'].sum().plot.bar(title = 'Lucro x Marca')
plt.xlabel('Marca')
plt.ylabel('Receita')
plt.xticks(rotation = 'horizontal')
plt.show()
```

```python
# Grafico com Lucro por Classe em 2009
df.groupby([df['Classe']])['Lucro'].sum().plot.bar(title = 'Lucro x Classe')
plt.xlabel('Classe')
plt.ylabel('Receita')
plt.xticks(rotation = 'horizontal')
plt.show()
```

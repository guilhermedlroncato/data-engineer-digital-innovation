```python
def valida_placa(placa):
    invalido = 0

    if '-' in placa:
        if (placa[0:3].isupper()) and (len(placa.split('-')[1]) == 4) and (len(placa.split('-')[0]) == 3):
            return True
        else:
            return False
    else:
        return False

n = int(input())
while n:
    n -= 1
    placa = input()
    if valida_placa(placa):
        try:            
            r = int(placa.split('-')[1][-1])
            if r <= 2 and r > 0:
                print('SEGUNDA')
            elif r <= 4 and r > 0:
                print('TERCA')
            elif r <= 6 and r > 0:
                print('QUARTA')
            elif r <= 8 and r > 0:
                print('QUINTA')
            else:
                print('SEXTA')
        except:
            print('FALHA')
    else:
        print('FALHA')
```
